```python
features_scaled = preprocessing.scale(features)
```

```python
labels_scaled = preprocessing.scale(newlabel)
```

```python
nrow = len(labels_scaled)
nrow
```

```python
import numpy as np
crossval = 10
for ival in range(crossval):
    idxTest = [a for a in range(nrow) if a%crossval == ival%crossval]
    idxTrain = [a for a in range(nrow) if a%crossval != ival%crossval]
    xTrain = np.array([features_scaled[r] for r in idxTrain])
    xTest = np.array([features_scaled[r] for r in idxTest])
    labelTrain = np.array([labels_scaled[r] for r in idxTrain])
    labelTest = np.array([labels_scaled[r] for r in idxTest])
    alphas, coefs, _ = enet_path(xTrain,labelTrain, l1_ratio = 0.8,fit_intercept = False, return_models = False)
    
    if ival == 0:
        pred = np.dot(xTest,coefs)
        yOut = labelTest
    else:
        #accumulate actual labels
        yTemp = np.array(yOut)
        yOut = np.concatenate((yTemp,labelTest),axis=0)
        
        #accumulate predictions
        
        predTemp = np.array(pred)
        pred = np.concatenate((predTemp,np.dot(xTest, coefs)), axis = 0)
```

## Let us analyse the misclassification rate with respect to different values of alph and get the optimum value of alpha

```python
misClassRate = []
for iPred in range(0,nPred):
    predList = list(pred[:,iPred])
    errCnt = 0.0
    for irow in range(nrow):
        if (predList[irow] < 0.0) and (yOut[irow] >= 0.0) :
            errCnt += 1.0
        elif (predList[irow] >= 0.0) and (yOut[irow] < 0.0):
            errCnt +=1.0
    misClassRate.append(errCnt/nrow)
```

```python
minError = min(misClassRate)
```

```python
minError
```

```python
idxMin = misClassRate.index(minError)
```

```python
idxMin
```

```python
plotAlphas = list(alphas[0:len(alphas)])
```

```python
plt.figure()
plt.plot(plotAlphas, misClassRate, label = 'Misclassification Error Across Folds', linewidth = 2)
plt.axvline(plotAlphas[idxMin], linestyle = '--', label = 'CV Estimate of best alpha')
plt.legend()
plt.semilogx()
ax = plt.gca()
ax.invert_xaxis()
plt.xlabel('alpha')
plt.ylabel('Misclassification Error')
plt.axis('tight')
plt.show()
```

## Let us analyse AUC and find out the value of alpha giving the maximum AUC

```python
idxpos = [i for i in range(nrow) if yOut[i] > 0.0]
yOutBin = [0] * nrow
for i in idxpos : yOutBin[i] = 1
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
auc = []
for iPred in range(0, nPred) :
    predList = list(pred[:, iPred])
    aucCalc = roc_auc_score(yOutBin, predList)
    auc.append(aucCalc)
```

```python
maxAUC = max(auc)
idxMax = auc.index(maxAUC)
```

```python
plt.figure()
plt.plot(plotAlphas, auc, label = 'AUC Across Folds', linewidth = 2)
plt.axvline(plotAlphas[idxMax], linestyle = '--', label = 'CV Estimate of best alpha')
plt.legend()
plt.semilogx()
ax = plt.gca()
ax.invert_xaxis()
plt.xlabel('alpha')
plt.ylabel('Area under the ROC curve')
plt.axis('tight')
plt.show()
```

```python
fpr, tpr, thresh = roc_curve(yOutBin, list(pred[:,idxMax]))
ctClass = [i * 0.01 for i in range(101)]
plt.plot(fpr, tpr, linewidth = 2)
plt.plot(ctClass, ctClass, linestyle = ':')
plt.xlabel('False positive Rate')
plt.ylabel('True positive Rate')
plt.show()
```
